# 第8章　特征值与特征向量（直觉篇） —— 抓住主要矛盾

> **本章目标**：理解 **A**·**v** = λ·**v** 的几何含义，建立 PCA 降维直觉。
> **前置知识**：第 6 章（矩阵乘法）

```python
import numpy as np
import matplotlib.pyplot as plt
print("✅ 环境就绪")
```

---
## 8.1　定义——矩阵作用下方向不变的向量

📐 **定义**：`**A**·**v** = λ·**v**`。**v** 被 **A** 乘了之后方向完全不变，只被缩放了 λ 倍。

```python
A=np.array([[4,1],[2,3]]); eigvals,eigvecs=np.linalg.eig(A)
for i in range(2):
    v=eigvecs[:,i]; lam=eigvals[i]
    print(f"λ{i+1}={lam:.3f}  A@v={A@v}  λ*v={lam*v}  匹配:{np.allclose(A@v,lam*v)}")
```

---
## 8.2　几何可视化——单位圆→椭圆的拉伸轴

```python
A=np.array([[2,0.5],[0.5,1]]); eigvals,eigvecs=np.linalg.eig(A)
fig,ax=plt.subplots(figsize=(6,6))
theta=np.linspace(0,2*np.pi,100)
circle=np.column_stack([np.cos(theta),np.sin(theta)])
transformed=circle@A.T
ax.plot(circle[:,0],circle[:,1],'b-',alpha=0.3,label='unit circle')
ax.plot(transformed[:,0],transformed[:,1],'r-',alpha=0.5,label='ellipse')
for i in range(2):
    v=eigvecs[:,i]; lam=eigvals[i]
    ax.arrow(0,0,v[0]*lam*2,v[1]*lam*2,head_width=0.15,color=['green','orange'][i],label=f'λ{i+1}={lam:.2f}')
ax.set_xlim(-5,5); ax.set_ylim(-5,5); ax.set_aspect('equal'); ax.legend()
ax.set_title('特征向量 = 椭圆的拉伸轴方向'); plt.show()
```

---
## 8.3　特征值的含义——信息的"浓度"

- **大特征值** → 该方向数据方差大 → "信息丰富"
- **小特征值** → 数据在该方向几乎不变 → 可丢弃（降维）

```python
np.random.seed(42); B=np.random.randn(5,5); A_sym=B.T@B
eigvals=np.sort(np.linalg.eigvalsh(A_sym))[::-1]
print("特征值  占比  累计")
for i,lam in enumerate(eigvals):
    r=lam/eigvals.sum()*100; cum=eigvals[:i+1].sum()/eigvals.sum()*100
    print(f"λ{i+1}={lam:.2f}  {r:5.1f}%  {cum:5.1f}%")
print(f"前2个解释了 {eigvals[:2].sum()/eigvals.sum()*100:.0f}% 的信息")
```
