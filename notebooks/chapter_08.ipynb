## 8.1-8.4  定义/求解/几何/PCA动机

Av=lambda*v。eig求解+排序+能量占比。单位圆经A变椭圆:轴=特征向量,半轴=特征值。前k大特征值=数据主方向。
> 桥梁提示：可先跳过eig细节,只记直觉,第17章图像理解。

AI连接：PCA(Ch17)=取前k大特征值方向投影;SVD(Ch9)=推广到任意矩阵。


```python
import numpy as np
A=np.array([[4.,1.],[1.,3.]])
evals,evecs=np.linalg.eig(A)
for i in range(2):
  v=evecs[:,i]; lam=evals[i]
  print(f'A@v={A@v.round(3)} lambda*v={(lam*v).round(3)} match={np.allclose(A@v,lam*v)}')
# Energy proportion
M=np.random.randn(5,5); A5=M.T@M; ev5=np.linalg.eig(A5)[0].real
ev5s=sorted(ev5,reverse=True); total=sum(ev5s)
for i,lam in enumerate(ev5s):
  print(f'lambda{i+1}={lam:.1f} ({lam/total*100:.0f}%)')
```
