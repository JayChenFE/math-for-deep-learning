```python
import numpy as np
A=np.array([[4.,1.],[1.,3.]])
evals,evecs=np.linalg.eig(A)
for i in range(2):
  v=evecs[:,i]; lam=evals[i]
  print('A@v={}, lambda*v={}, match={}'.format(A@v.round(3),(lam*v).round(3),np.allclose(A@v,lam*v)))
M=np.random.randn(5,5); A5=M.T@M
ev5=np.linalg.eig(A5)[0].real
ev5s=sorted(ev5,reverse=True)
total=sum(ev5s)
for i,lam in enumerate(ev5s):
  print('lambda{}={:.1f} ({:.0f}%)'.format(i+1,lam,lam/total*100))
```

```python
import numpy as np, matplotlib.pyplot as plt
A=np.array([[4.,1.],[1.,3.]])
th=np.linspace(0,2*np.pi,200)
circle=np.column_stack([np.cos(th),np.sin(th)])
ellipse=(A@circle.T).T
evals,evecs=np.linalg.eig(A)
fig,ax=plt.subplots(figsize=(7,7))
ax.plot(circle[:,0],circle[:,1],'b-',alpha=0.5)
ax.plot(ellipse[:,0],ellipse[:,1],'r-',lw=2)
for i in range(2):
  v=evecs[:,i]*evals[i]
  ax.quiver(0,0,v[0],v[1],angles='xy',scale_units='xy',scale=1,color=['green','orange'][i],width=0.03)
ax.set_aspect('equal'); ax.set_xlim(-6,6); ax.set_ylim(-6,6)
ax.set_title('unit circle -> ellipse (axes = eigenvectors)')
print('ellipse axes = eigenvector directions')
```
