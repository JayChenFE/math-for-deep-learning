## 17.1-17.5  期望/方差/协方差/PCA从零实现

E[X]=数据中心。Var[X]=离散度(std=sqrt(Var))。Cov>0正相关/<0负相关。协方差矩阵特征分解->特征向量=数据主方向(PC)。取前k投影=降维。

AI连接：BatchNorm(Ch21)用(x-mu)/sigma,权重初始化(Ch25)保持方差一致。


```python
import numpy as np
np.random.seed(42)
t=np.random.normal(5,1,1000); w=np.random.normal(5,5,1000)
print(f'tight: mean={t.mean():.3f} std={t.std():.3f}')
print(f'wide:  mean={w.mean():.3f} std={w.std():.3f}')
# Covariance
h=np.random.normal(170,10,500); wt=0.8*h+np.random.normal(0,8,500)
print(f'cov(H,W)={np.cov(h,wt)[0,1]:.1f} corr={np.corrcoef(h,wt)[0,1]:.3f}')
# PCA from scratch
n=50; c0=np.random.randn(n,4)*0.5+np.array([2,3,1,0])
c1=np.random.randn(n,4)*0.6+np.array([-1,-1,0,2])
c2=np.random.randn(n,4)*0.4+np.array([3,-2,2,-1])
X=np.vstack([c0,c1,c2]); Xc=X-X.mean(axis=0)
cov_mat=np.cov(Xc,rowvar=False)
evals,evecs=np.linalg.eig(cov_mat); evals=evals.real; evecs=evecs.real
idx=np.argsort(evals)[::-1]; W_pca=evecs[:,idx[:2]]; Xp=Xc@W_pca
total=evals.sum()
for i in range(4): print(f'PC{i+1}: {evals[idx[i]]:.2f} ({evals[idx[i]]/total*100:.0f}%)')
print(f'First 2 PCs retain {evals[idx[:2]].sum()/total*100:.0f}% variance')
```
