```python
import numpy as np
np.random.seed(42)
t=np.random.normal(5,1,1000); w=np.random.normal(5,5,1000)
print("tight: mean={:.3f} std={:.3f}".format(t.mean(),t.std()))
print("wide:  mean={:.3f} std={:.3f}".format(w.mean(),w.std()))
# Covariance & correlation
h=np.random.normal(170,10,500); wt=0.8*h+np.random.normal(0,8,500)
print("cov(height,weight)={:.1f} corr={:.3f}".format(np.cov(h,wt)[0,1],np.corrcoef(h,wt)[0,1]))
tmp=np.random.normal(15,8,500); ht=-3*tmp+np.random.normal(0,10,500)
print("cov(temp,heating)={:.1f} corr={:.3f}".format(np.cov(tmp,ht)[0,1],np.corrcoef(tmp,ht)[0,1]))
```

```python
import numpy as np, matplotlib.pyplot as plt
np.random.seed(42)
n=50
c0=np.random.randn(n,4)*0.5+np.array([2,3,1,0])
c1=np.random.randn(n,4)*0.6+np.array([-1,-1,0,2])
c2=np.random.randn(n,4)*0.4+np.array([3,-2,2,-1])
X=np.vstack([c0,c1,c2]); y=np.array([0]*n+[1]*n+[2]*n)
Xc=X-X.mean(axis=0)
cov_mat=np.cov(Xc,rowvar=False)
evals,evecs=np.linalg.eig(cov_mat); evals=evals.real; evecs=evecs.real
idx=np.argsort(evals)[::-1]; W=evecs[:,idx[:2]]
Xp=Xc@W  # (150,4)@(4,2)->(150,2)
total=evals.sum()
for i in range(4): print("PC{}: {:.2f} ({:.0f}%)".format(i+1,evals[idx[i]],evals[idx[i]]/total*100))
print("First 2 PCs retain {:.0f}% variance".format(evals[idx[:2]].sum()/total*100))
fig,ax=plt.subplots(figsize=(7,5))
for c in range(3):
  m=y==c; ax.scatter(Xp[m,0],Xp[m,1],c=["red","green","blue"][c],alpha=0.6,s=15,label="Class {}".format(c))
ax.set_xlabel("PC1"); ax.set_ylabel("PC2"); ax.set_title("PCA: 4D->2D"); ax.legend(); plt.show()
```
